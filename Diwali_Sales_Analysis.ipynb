```python
# import python libraries

import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt # visualizing data
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv(r'C:\Users\sanja\Data analytics\Project\Diwali Sales Data.csv', encoding='unicode_escape')
```

```python
#drop unrelated/blank columns
df.drop(['Status', 'unnamed1'], axis=1, inplace=True)
```

```python
#check for null values
pd.isnull(df).sum()
```

```python
# drop null values
df.dropna(inplace=True)
```

```python
# change data type
df['Amount'] = df['Amount'].astype('int')
```

```python
df['Amount'].dtypes
```

```python
df.columns
```

```python
#rename column
df.rename(columns= {'Marital_Status':'Shaadi'})
```

```python
# describe() method returns description of the data in the DataFrame (i.e. count, mean, std, etc)
df.describe()
```

```python
#use descibe() for sepecific columns
df[['Age','Orders','Amount']].describe()
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
# plotting a bar chart for Gender and it's count

ax = sns.countplot(x = 'Gender',data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# plotting a bar chart for gender vs total amount

sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Gender',y= 'Amount' ,data = sales_gen, color='orange')
```

```python
# plotting a bar chart for Gender and it's count

ax = sns.countplot(x = 'Gender',data = df,hue= 'Gender', palette={'M': 'skyblue', 'F': 'lightcoral'})#color change 

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
ax = sns.countplot(data = df, x = 'Age Group', hue = 'Gender', palette={'M': 'skyblue', 'F': 'red'})

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Total Amount vs Age Group
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
# Set the figure size
sns.set(rc={'figure.figsize': (25, 5)})
# Set the background color
sns.set_theme(style="darkgrid", palette="pastel")
# Create the bar plot with different colors for each age group
ax = sns.barplot(x='Age Group', y='Amount', data=sales_age,hue='Age Group', palette='viridis')

# Add labels to the bars
for bars in ax.containers:
    ax.bar_label(bars, fmt='%.2f', label_type='edge')

# Show the plot
plt.show()
```

```python
# total number of orders from top 10 states

sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.set_theme(style="whitegrid",palette="pastel")
sns.barplot(data = sales_state, x = 'State',y= 'Orders')
```

```python
# total amount/sales from top 10 states

sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
# Set the background color
sns.set_theme(style="whitegrid", palette="pastel")
# Create the bar plot with different colors for each state
ax = sns.barplot(data=sales_state, x='State', y='Amount',hue='State', palette='viridis')

# Add labels to the bars
for bars in ax.containers:
    ax.bar_label(bars, fmt='%.2f', label_type='edge')
plt.show
```

```python
sales_state = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data = sales_state, x = 'Marital_Status',y= 'Amount', hue='Gender')
```

```python
# Set figure size
sns.set(rc={'figure.figsize': (20, 5)})

# Set background style
sns.set_style("whitegrid")  # You can choose other styles like "darkgrid", "white", etc.

# Create the countplot with different color for each bar
ax = sns.countplot(data=df, x='Occupation',hue='Occupation', palette='Set2',legend=False)
ax.set_ylabel('Count', fontsize=14, fontweight='bold')
ax.set_xlabel('Occupation', fontsize=14, fontweight='bold')
# Add labels to the bars
for bars in ax.containers:
    ax.bar_label(bars, fmt='%d', label_type='edge', fontsize=12, color='black', fontweight='bold')

# Change font family
plt.rcParams['font.family'] = 'sans-serif'  # Change to the desired font family

# Show the plot
plt.show()
```

```python
# Group by Occupation and calculate total Amount
sales_state = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

# Set figure size
sns.set(rc={'figure.figsize': (20, 5)})

# Set background style
sns.set_style("whitegrid")  # You can choose other styles like "darkgrid", "white", etc.

# Create the bar plot with customizations
ax = sns.barplot(data=sales_state, x='Occupation', y='Amount', hue='Occupation',palette='viridis')

# Customize x-axis labels
#ax.set_xticklabels(ax.get_xticklabels(), rotation=45, horizontalalignment='right', fontsize=12, fontweight='bold')
ax.set_xlabel('Occupation', fontsize=14, fontweight='bold')
# Customize y-axis label
ax.set_ylabel('Total Amount', fontsize=14, fontweight='bold')

# Add labels to the bars with customizations
for bars in ax.containers:
    ax.bar_label(bars, fmt='%d', label_type='edge', fontsize=12, color='black', fontweight='bold')

# Change font family
plt.rcParams['font.family'] = 'sans-serif'  # Change to the desired font family

# Set background color
ax.set_facecolor('#f0f0f0')  # Change to the desired background color

# Show the plot
plt.show()
```

```python
#sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

#sns.set(rc={'figure.figsize':(20,5)})
#sns.barplot(data = sales_state, x = 'Product_Category',y= 'Amount')

sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

# Set figure size
sns.set(rc={'figure.figsize': (20, 5)})

# Set background style
sns.set_style("whitegrid")

# Create the barplot with customizations
ax = sns.barplot(data=sales_state, x='Product_Category', y='Amount',palette='viridis')

# Customize x-axis labels
#ax.set_xticklabels(ax.get_xticklabels(), rotation=0, horizontalalignment='right', fontsize=12)
ax.set_xlabel('Product_Category', fontsize=14, fontweight='bold',rotation=0, horizontalalignment='right')
# Customize y-axis label
ax.set_ylabel('Total Amount', fontsize=14, fontweight='bold')

# Add data labels to the bars with customizations
for bars in ax.containers:
    ax.bar_label(bars, fmt='%.2f', label_type='edge', fontsize=12, color='black', fontweight='bold')

# Set title
ax.set_title('Top 10 Product Categories by Total Amount', fontsize=16, fontweight='bold')

# Change font family
plt.rcParams['font.family'] = 'sans-serif'

# Set background color
ax.set_facecolor('#f0f0f0')

# Show the plot
plt.show()
```

```python
#sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

#sns.set(rc={'figure.figsize':(20,5)})
#sns.barplot(data = sales_state, x = 'Product_ID',y= 'Orders')

# Group by Product_ID and calculate total Orders, then sort and select top 10
sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

# Set figure size
sns.set(rc={'figure.figsize': (20, 5)})

# Set background style
sns.set_style("whitegrid")

# Create the barplot with customizations
ax = sns.barplot(data=sales_state, x='Product_ID', y='Orders', hue='Product_ID', palette='plasma')

# Customize x-axis labels
#ax.set_xticklabels(ax.get_xticklabels(), rotation=45, horizontalalignment='right', fontsize=12, fontweight='bold')
ax.set_xlabel('Product_ID', fontsize=14, fontweight='bold')
# Customize y-axis label
ax.set_ylabel('Total Orders', fontsize=14, fontweight='bold')

# Add data labels to the bars with customizations
for bars in ax.containers:
    ax.bar_label(bars, fmt='%d', label_type='edge', fontsize=12, color='black', fontweight='bold')

# Set title
ax.set_title('Top 10 Product IDs by Total Orders', fontsize=16, fontweight='bold')

# Change font family
plt.rcParams['font.family'] = 'sans-serif'

# Set background color
ax.set_facecolor('#f0f0f0')

# Show the plot
plt.show()

```

```python
# top 10 most sold products (same thing as above)

#fig1, ax1 = plt.subplots(figsize=(12,7))
#df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')
import random
# Group by Product_ID and calculate total Orders, then select top 10
top_products = df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False)

# Set figure size
sns.set(rc={'figure.figsize': (12, 7)})

# Set background style
sns.set_style("whitegrid")

# Create a list of random colors for each bar
bar_colors = [f'#{random.randint(0, 0xFFFFFF):06x}' for _ in range(len(top_products))]

# Create the barplot with customizations
ax = top_products.plot(kind='bar', color=bar_colors)

# Customize x-axis labels
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, horizontalalignment='right', fontsize=12, fontweight='bold')

# Customize y-axis label
ax.set_ylabel('Total Orders', fontsize=14, fontweight='bold')

# Add data labels to the bars with customizations
for bars in ax.containers:
    ax.bar_label(bars, fmt='%d', label_type='edge', fontsize=12, color='black', fontweight='bold')

# Set title
ax.set_title('Top 10 Most Sold Products', fontsize=16, fontweight='bold')

# Change font family
plt.rcParams['font.family'] = 'sans-serif'

# Set background color
ax.set_facecolor('#f0f0f0')

# Show the plot
plt.show()
```
